Let's test connection using our library

```python
from urllib import response
from ibm_wmla_client import Connection
import numpy as np
import os
```

```python
# diable warning (unverified HTTPS request)
# please refer: https://urllib3.readthedocs.io/en/1.26.x/advanced-usage.html#ssl-warnings

import urllib3
urllib3.disable_warnings()
```

Define variables for setup WMLA environment (we will use USER_ACCESS_TOKEN to connect WMLA service)

```python
service_url = "https://wmla-console-wmla.apps.ocp.tanggle.com"
service_instance = 'ibm_wml_elastic_distributed_inference'

USER_ACCESS_TOKEN = os.getenv('USER_ACCESS_TOKEN')
```

```python
edi_connection = Connection(service_url, service_instance, wmla_v1=True, edi=True,
                 apikey=None, username=None, password=None, user_access_token = USER_ACCESS_TOKEN)
```

```python
edi_connection.connect()
```

```python
conn = edi_connection.service_edi
```

get model's list and its information

```python
print(conn.get_models(verify=False))
```

get details of the model (pingpong)

```python
print(conn.get_model("pingpong"))
```
